## crosstab

```python
import pandas as pd
df = pd.read_excel(r"C:\Users\Admin\Desktop\Preprocessing\Pandas\tutorial\survey.xls")
df
```

```python
pd.crosstab(df.Nationality,df.Handedness)
```

```python
pd.crosstab(df.Handedness,df.Nationality)
```

```python
pd.crosstab(df.Sex,df.Handedness)
```

## Margins (all sum)

```python
pd.crosstab(df.Sex,df.Handedness, margins=True)
```

## Multi Index Column and Rows

```python
pd.crosstab(df.Sex,[df.Handedness,df.Nationality])
```

```python
pd.crosstab(df.Sex, [df.Handedness,df.Nationality], margins=True)
```

```python
pd.crosstab([df.Nationality,df.Sex],[df.Handedness])
```

```python
pd.crosstab([df.Nationality, df.Sex], [df.Handedness], margins=True)
```

## Normalize (%)

```python
pd.crosstab(df.Sex, df.Handedness,normalize='index')
```

## Aggfunc and Values

```python
import numpy as np  
pd.crosstab(df.Sex, df.Handedness,values=df.Age,aggfunc=np.average)
```

```python
pd.crosstab(df.Sex, df.Handedness,values=df.Age,aggfunc=np.sum)
```

```python
pd.crosstab(df.Nationality, df.Handedness, values=df.Age, aggfunc=np.average)
```

```python
pd.crosstab(df.Nationality, df.Sex, values= df.Age, aggfunc= np.sum)
```
